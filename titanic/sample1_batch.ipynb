```python
#https://www.kaggle.com/arthurtok/introduction-to-ensembling-stacking-in-python
##検索用文字列
#テスト分割する場合

# Load in our libraries
import pandas as pd
import numpy as np
import re
import sklearn
import xgboost as xgb
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.tools as tls

import warnings
warnings.filterwarnings('ignore')

# Going to use these 5 base models for the stacking
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier, ExtraTreesClassifier
from sklearn.svm import SVC, SVR, LinearSVC
from sklearn.cross_validation import KFold;
from sklearn.cross_validation import train_test_split
# Load in the train and test datasets
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

# Store our passenger ID for easy access
PassengerId = test['PassengerId']
#dataset
#http://www.mirandora.com/?p=1804

#参考
#https://www.kaggle.com/sinakhorami/titanic/titanic-best-working-classifier

full_data = [train, test]

# Some features of my own that I have added in
# Gives the length of the name
train['Name_length'] = train['Name'].apply(len)
test['Name_length'] = test['Name'].apply(len)
# Feature that tells whether a passenger had a cabin on the Titanic　[Cabin=船室]
train['Has_Cabin'] = train["Cabin"].apply(lambda x: 0 if type(x) == float else 1)
test['Has_Cabin'] = test["Cabin"].apply(lambda x: 0 if type(x) == float else 1)

# Feature engineering steps taken from Sina
# Create new feature FamilySize as a combination of SibSp and Parch
#[sibsp	乗船している夫婦、兄弟姉妹の数, parch	乗船している親、子供の数]
for dataset in full_data:
    dataset['FamilySize'] = dataset['SibSp'] + dataset['Parch'] + 1
# Create new feature IsAlone from FamilySize
for dataset in full_data:
    dataset['IsAlone'] = 0
    dataset.loc[dataset['FamilySize'] == 1, 'IsAlone'] = 1
# Remove all NULLS in the Embarked column
for dataset in full_data:
    dataset['Embarked'] = dataset['Embarked'].fillna('S')
# Remove all NULLS in the Fare column and create a new feature CategoricalFare
for dataset in full_data:
    dataset['Fare'] = dataset['Fare'].fillna(train['Fare'].median())
train['CategoricalFare'] = pd.qcut(train['Fare'], 4)
# Create a New feature CategoricalAge
for dataset in full_data:
    age_avg = dataset['Age'].mean()
    age_std = dataset['Age'].std()
    age_null_count = dataset['Age'].isnull().sum()
    age_null_random_list = np.random.randint(age_avg - age_std, age_avg + age_std, size=age_null_count)
    dataset['Age'][np.isnan(dataset['Age'])] = age_null_random_list
    dataset['Age'] = dataset['Age'].astype(int)
train['CategoricalAge'] = pd.cut(train['Age'], 5)
# Define function to extract titles from passenger names
def get_title(name):
    title_search = re.search(' ([A-Za-z]+)\.', name)
    # If the title exists, extract and return it.
    if title_search:
        return title_search.group(1)
    return ""
# Create a new feature Title, containing the titles of passenger names
for dataset in full_data:
    dataset['Title'] = dataset['Name'].apply(get_title)
# Group all non-common titles into one single grouping "Rare"
for dataset in full_data:
    dataset['Title'] = dataset['Title'].replace(['Lady', 'Countess','Capt', 'Col','Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')

    dataset['Title'] = dataset['Title'].replace('Mlle', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Ms', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Mme', 'Mrs')

for dataset in full_data:
    # Mapping Sex
    dataset['Sex'] = dataset['Sex'].map( {'female': 0, 'male': 1} ).astype(int)
    
    # Mapping titles
    title_mapping = {"Mr": 1, "Miss": 2, "Mrs": 3, "Master": 4, "Rare": 5}
    dataset['Title'] = dataset['Title'].map(title_mapping)
    dataset['Title'] = dataset['Title'].fillna(0)
    
    # Mapping Embarked
    #[embarked	乗船場所(C = Cherbourg; Q = Queenstown; S = Southampton)]
    dataset['Embarked'] = dataset['Embarked'].map( {'S': 0, 'C': 1, 'Q': 2} ).astype(int)
    
    # Mapping Fare
    dataset.loc[ dataset['Fare'] <= 7.91, 'Fare'] 						        = 0
    dataset.loc[(dataset['Fare'] > 7.91) & (dataset['Fare'] <= 14.454), 'Fare'] = 1
    dataset.loc[(dataset['Fare'] > 14.454) & (dataset['Fare'] <= 31), 'Fare']   = 2
    dataset.loc[ dataset['Fare'] > 31, 'Fare'] 							        = 3
    dataset['Fare'] = dataset['Fare'].astype(int)
    
    # Mapping Age
    dataset.loc[ dataset['Age'] <= 16, 'Age'] 					       = 0
    dataset.loc[(dataset['Age'] > 16) & (dataset['Age'] <= 32), 'Age'] = 1
    dataset.loc[(dataset['Age'] > 32) & (dataset['Age'] <= 48), 'Age'] = 2
    dataset.loc[(dataset['Age'] > 48) & (dataset['Age'] <= 64), 'Age'] = 3
    dataset.loc[ dataset['Age'] > 64, 'Age'] ;

#--------------------------------------------------------------------------------------------
# Feature selection
drop_elements = ['PassengerId', 'Name', 'Ticket', 'Cabin', 'SibSp']
train = train.drop(drop_elements, axis = 1)
train = train.drop(['CategoricalAge', 'CategoricalFare'], axis = 1)
print(type(train))
train["Pclass"] = np.log(train["Pclass"]+1.1)
train["Sex"] = np.log(train["Sex"]+1.1)
train["Age"] = np.log(train["Age"]+1.1)
train["Parch"] = np.log(train["Parch"]+1.1)
train["Fare"] = np.log(train["Fare"]+1.1)
train["Embarked"] = np.log(train["Embarked"]+1.1)
train["Name_length"] = np.log(train["Name_length"]+1.1)
train["Has_Cabin"] = np.log(train["Has_Cabin"]+1.1)
train["FamilySize"] = np.log(train["FamilySize"]+1.1)
train["IsAlone"] = np.log(train["IsAlone"]+1.1)
train["Title"] = np.log(train["Title"]+1.1)

#print(numpy.log(train))
test  = test.drop(drop_elements, axis = 1)
test["Pclass"] = np.log(test["Pclass"]+1.1)
test["Sex"] = np.log(test["Sex"]+1.1)
test["Age"] = np.log(test["Age"]+1.1)
test["Parch"] = np.log(test["Parch"]+1.1)
test["Fare"] = np.log(test["Fare"]+1.1)
test["Embarked"] = np.log(test["Embarked"]+1.1)
test["Name_length"] = np.log(test["Name_length"]+1.1)
test["Has_Cabin"] = np.log(test["Has_Cabin"]+1.1)
test["FamilySize"] = np.log(test["FamilySize"]+1.1)
test["IsAlone"] = np.log(test["IsAlone"]+1.1)
test["Title"] = np.log(test["Title"]+1.1)
# print(test)

#--------------------------------------------------------------------------------------------
# Create Numpy arrays of train, test and target ( Survived) dataframes to feed into our models
y_train = train['Survived'].ravel()
train = train.drop(['Survived'], axis=1)
x_train = train.values # Creates an array of the train data
x_test = test.values # Creats an array of the test data


TEST_SIZE = 0.2
RANDOM_STATE = 0
y_test=np.zeros(len(x_test))

print("x_test ", len(x_test))
print("y_test ", len(y_test))

x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=TEST_SIZE, random_state=RANDOM_STATE) #テスト分割する場合
x_test, x_test2, y_test, y_test2 = train_test_split(x_test, y_test, test_size=TEST_SIZE, random_state=RANDOM_STATE) #テスト分割する場合
#--------------------------------------------------------------------------------------------
#ラッパー作成

# Some useful parameters which will come in handy later on
# ntrain = train.shape[0]
# ntest = test.shape[0] 
ntrain = x_train.shape[0] #テスト分割する場合
ntest = x_test.shape[0] #テスト分割する場合
SEED = 0 # for reproducibility
NFOLDS = 5 # set folds for out-of-fold prediction
kf = KFold(ntrain, n_folds= NFOLDS, random_state=SEED)

# Class to extend the Sklearn classifier
class SklearnHelper(object):
    def __init__(self, clf, seed=0, params=None):
        if seed != -1:
            params['random_state'] = seed
        self.clf = clf(**params)

    def train(self, x_train, y_train):
        self.clf.fit(x_train, y_train)

    def predict(self, x):
        return self.clf.predict(x)
    
    def fit(self,x,y):
        return self.clf.fit(x,y)
    
    def feature_importances(self,x,y):
        print(self.clf.fit(x,y).feature_importances_)
    
# Class to extend XGboost classifer
#--------------------------------------------------------------------------------------------
# Put in our parameters for said classifiers
# Random Forest parameters
rf_params = {
    'n_jobs': -1,
    'n_estimators': 500,
     'warm_start': True, 
     #'max_features': 0.2,
    'max_depth': 6,
    'min_samples_leaf': 2,
    'max_features' : 'sqrt',
    'verbose': 0
}

# Extra Trees Parameters
et_params = {
    'n_jobs': -1,
    'n_estimators':500,
    #'max_features': 0.5,
    'max_depth': 8,
    'min_samples_leaf': 2,
    'verbose': 0
}

# AdaBoost parameters
ada_params = {
    'n_estimators': 500,
    'learning_rate' : 0.75
}

# Gradient Boosting parameters
gb_params = {
    'n_estimators': 500,
     #'max_features': 0.2,
    'max_depth': 5,
    'min_samples_leaf': 2,
    'verbose': 0
}

# Support Vector Classifier parameters 線形
svc_linear_params = {
    'kernel' : 'linear',
    'C' : 0.025
    }

# Support Vector Classifer parameters 多項式カーネル
svc_poly_params = {
    'kernel' : 'poly',
    'degree' : 3,
    'C' : 0.025
    }

# Support Vector Classifer parameters RBF カーネル
svc_rbf_params = {
    'kernel' : 'rbf'
    }

# Support Vector Classifer parameters sigmoid
svc_sigmoid_params = {
    'kernel' : 'sigmoid'
    }

# Support Vector Classifer parameters 線形
svr_linear_params = {
    'kernel' : 'linear',
    'C' : 0.5
    }

# Linear Support Vector Classifer parameters 線形
linear_svc_params = {
    'C' : 1
    }


#ここ
#--------------------------------------------------------------------------------------------
def get_oof(clf, x_train, y_train, x_test):
    oof_train = np.zeros((ntrain,))
    oof_test = np.zeros((ntest,))
    oof_test_skf = np.empty((NFOLDS, ntest))

    for i, (train_index, test_index) in enumerate(kf):
        x_tr = x_train[train_index]
        y_tr = y_train[train_index]
        x_te = x_train[test_index]

        clf.train(x_tr, y_tr)

        oof_train[test_index] = clf.predict(x_te)
        oof_test_skf[i, :] = clf.predict(x_test)

    oof_test[:] = oof_test_skf.mean(axis=0)
    return oof_train.reshape(-1, 1), oof_test.reshape(-1, 1)
#--------------------------------------------------------------------------------------------
# Create 5 objects that represent our 4 models
rf = SklearnHelper(clf=RandomForestClassifier, seed=SEED, params=rf_params)
et = SklearnHelper(clf=ExtraTreesClassifier, seed=SEED, params=et_params)
ada = SklearnHelper(clf=AdaBoostClassifier, seed=SEED, params=ada_params)
gb = SklearnHelper(clf=GradientBoostingClassifier, seed=SEED, params=gb_params)
svc_linear = SklearnHelper(clf=SVC, seed=SEED, params=svc_linear_params)
svc_poly = SklearnHelper(clf=SVC, seed=SEED, params=svc_poly_params)
svc_rbf = SklearnHelper(clf=SVC, seed=SEED, params=svc_rbf_params)
svc_sigmoid = SklearnHelper(clf=SVC, seed=SEED, params=svc_sigmoid_params)
svr_linear = SklearnHelper(clf=SVR, seed=-1, params=svr_linear_params)
linear_svc = SklearnHelper(clf=LinearSVC, seed=SEED, params=linear_svc_params)
# svc_precomputed = SklearnHelper(clf=SVC, seed=SEED, params=svc_precomputed_params)


#ここ
#--------------------------------------------------------------------------------------------
# # Create Numpy arrays of train, test and target ( Survived) dataframes to feed into our models
# y_train = train['Survived'].ravel()
# train = train.drop(['Survived'], axis=1)
# x_train = train.values # Creates an array of the train data
# x_test = test.values # Creats an array of the test data

# TEST_SIZE = 0.2
# RANDOM_STATE = 0
# y_test=np.zeros(len(x_test))

# print("x_test ", len(x_test))
# print("y_test ", len(y_test))

# x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=TEST_SIZE, random_state=RANDOM_STATE)
# x_test, x_test2, y_test, y_test2 = train_test_split(x_test, y_test, test_size=TEST_SIZE, random_state=RANDOM_STATE)
#--------------------------------------------------------------------------------------------
# Create our OOF train and test predictions. These base results will be used as new features

et_oof_train, et_oof_test = get_oof(et, x_train, y_train, x_test) # Extra Trees
rf_oof_train, rf_oof_test = get_oof(rf,x_train, y_train, x_test) # Random Forest
ada_oof_train, ada_oof_test = get_oof(ada, x_train, y_train, x_test) # AdaBoost 
gb_oof_train, gb_oof_test = get_oof(gb,x_train, y_train, x_test) # Gradient Boost
svc_linear_oof_train, svc_linear_oof_test = get_oof(svc_linear,x_train, y_train, x_test) # Support Vector Classifier
svc_poly_oof_train, svc_poly_oof_test = get_oof(svc_poly,x_train, y_train, x_test) # Support Vector Classifier
svc_rbf_oof_train, svc_rbf_oof_test = get_oof(svc_rbf,x_train, y_train, x_test) # Support Vector Classifier
svc_sigmoid_oof_train, svc_sigmoid_oof_test = get_oof(svc_sigmoid,x_train, y_train, x_test) # Support Vector Classifier
svr_linear_oof_train, svr_linear_oof_test = get_oof(svr_linear,x_train, y_train, x_test) # Support Vector Classifier
linear_svc_oof_train, linear_svc_oof_test = get_oof(linear_svc,x_train, y_train, x_test) # Support Vector Classifier


#ここ

print("Training is complete")

#--------------------------------------------------------------------------------------------
base_predictions_train = pd.DataFrame( {'RandomForest': rf_oof_train.ravel(),
     'ExtraTrees': et_oof_train.ravel(),
     'AdaBoost': ada_oof_train.ravel(),
      'GradientBoost': gb_oof_train.ravel()
    })
# base_predictions_train.head()
#--------------------------------------------------------------------------------------------
#ここ
x_train = np.concatenate(( et_oof_train, rf_oof_train, ada_oof_train, gb_oof_train, svc_linear_oof_train, svc_poly_oof_train, svc_rbf_oof_train, svc_sigmoid_oof_train, svr_linear_oof_train, linear_svc_oof_train), axis=1)
x_test = np.concatenate(( et_oof_test, rf_oof_test, ada_oof_test, gb_oof_test, svc_linear_oof_test, svc_poly_oof_test, svc_rbf_oof_test, svc_sigmoid_oof_test, svr_linear_oof_test, linear_svc_oof_test), axis=1)
#--------------------------------------------------------------------------------------------
gbm = xgb.XGBClassifier(
    #learning_rate = 0.02,
 n_estimators= 2000,
 max_depth= 4,
 min_child_weight= 2,
 #gamma=1,
 gamma=0.9,                        
 subsample=0.8,
 colsample_bytree=0.8,
 objective= 'binary:logistic',
 nthread= -1,
 scale_pos_weight=1).fit(x_train, y_train)
predictions = gbm.predict(x_test)

#--------------------------------------------------------------------------------------------
# Generate Submission File 
# StackingSubmission = pd.DataFrame({ 'PassengerId': PassengerId,
#                             'Survived': predictions })
# StackingSubmission.to_csv("StackingSubmission.csv", index=False)
#--------------------------------------------------------------------------------------------
print("end")
#--------------------------------------------------------------------------------------------
#--------------------------------------------------------------------------------------------
```
