```python
#!/usr/bin/env python
# -*- coding:utf-8 -*-
import numpy as np
import matplotlib.pyplot as plt
import pandas
from scipy.spatial import Voronoi, voronoi_plot_2d

npoint=10
maxvalue=10
points =np.array([[np.random.rand()*maxvalue,np.random.rand()*maxvalue] for i in range(npoint)])

vor = Voronoi(points)
voronoi_plot_2d(vor)
plt.show()
```
